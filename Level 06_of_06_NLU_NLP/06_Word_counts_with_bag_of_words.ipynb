------------
1.	Bag-of-words is a basic method for finding **topics** in a text
2.	Need to first create tokens using tokenization, and then count up all the tokens
3.	The more frequent a word, the more **important** it might be *
4.	It can be a great way to determine the significant words in a text

## Exercise 1: Building a Counter with bag-of-words

```python
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
counter = Counter(word_tokenize(
"""The cat is in the box. The cat likes the box. The box is over the cat."""))
```

```python
counter
```

```python
counter.most_common(5)
```

## Exercise 2: Building a Counter with bag-of-words and identify the topic

```python
#import os
#os.chdir("C:\\Users\\ramreddymyla\\Google Drive\\01 DS ML DL NLP and AI With Python Lab Copy\\02 Lab Data\\Python")
article = open('nlp_topic_identification.txt', 'r').read()
```

```python
article
```

```python
# Import Counter
from nltk.tokenize import word_tokenize
from collections import Counter
```

```python
# Tokenize the article: tokens
tokens = word_tokenize(article)
tokens
```

```python
# Convert the tokens into lowercase: lower_tokens
lower_tokens = [t.lower() for t in tokens]
lower_tokens
```

```python
# Create a Counter with the lowercase tokens: bow_simple
bow_simple = Counter(lower_tokens)
```

```python
# Print the 10 most common tokens
print(bow_simple.most_common(20))
```
