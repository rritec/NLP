```python
import nltk
document="""the little yellow dog barked at the cat"""

def ie_preprocess(document):
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences

sentence1=ie_preprocess(document)

grammar = "NP: {<DT>?<JJ>*<NN>}"
cp = nltk.RegexpParser(grammar)
result = cp.parse(sentence1[0])

#import os
#from IPython.display import Image
#nltk.draw.tree.TreeView(result)._cframe.print_to_file('output.ps')
#os.system('convert output.ps output.png')
#Image(filename='output.png') 
#from IPython.display import display
#display(result)

print(result)
result.draw()
```

## Exercise 2: NER with NLTK using scraped news article

```python
# Load a scraped news article
article = open('nlp_ner.txt', 'r').read()
```

```python
type(article)
```

```python
article
```

```python
# Tokenize the article into sentences: sentences
sentences = nltk.sent_tokenize(article)
```

```python
sentences
```

```python
len(sentences)
```

```python
sentences[0]
```

```python
# Tokenize each sentence into words: token_sentences
bow_sent = [nltk.word_tokenize(sent) for sent in sentences]
```

```python
bow_sent
```

```python
type(bow_sent)
```

```python
# Tag each tokenized sentence into parts of speech: pos_sentences
bow_pos = [nltk.pos_tag(w) for w in bow_sent]
```

```python
bow_pos
```

```python
# Create the named entity chunks: chunked_sentences
bow_pos_chunk = nltk.ne_chunk_sents(bow_pos, binary=True)
```

```python
bow_pos_chunk
```

```python
# Test for stems of the tree with 'NE' tags
for bag_of_chunks in bow_pos_chunk:
    for chunk in bag_of_chunks:
        if hasattr(chunk, "label") and chunk.label() == "NE":
            print(chunk)
```

```python
def fun_NE(filename):
    article = open(filename, 'r').read()
    sentences = nltk.sent_tokenize(article) # 1
    bow_sent = [nltk.word_tokenize(sent) for sent in sentences] #2
    bow_pos = [nltk.pos_tag(w) for w in bow_sent] #3
    bow_pos_chunk = nltk.ne_chunk_sents(bow_pos, binary=True) #4
    return bow_pos_chunk     
```

```python
a = fun_NE("nlp_ner.txt")

# Test for stems of the tree with 'NE' tags
for bag_of_chunks in a:
    for chunk in bag_of_chunks:
        #print(chunk)
        if hasattr(chunk, "label") and chunk.label() == "NE":
            print(chunk)
```
